```python
import json
import pickle as pkl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pymatgen.core import Composition, Element
import pprint as pp
```

```python
# with open('data/ss_extracted_imputed_precs_add_ops_all_targets.pkl', 'rb') as f:
#     full_papers = pkl.load(f)
# with open('data/sg_extracted_imputed_precs_add_ops_all_targets.pkl', 'rb') as f:
#     full_papers = pkl.load(f)
# with open('data/sg_extracted_imputed_precs_all_targets.pkl', 'rb') as f:
#     full_papers = pkl.load(f)
with open('data/ss_extracted_imputed_precs_all_targets.pkl', 'rb') as f:
    full_papers = pkl.load(f)
```

```python
len(full_papers)
```

```python
ss_targets = [x['target'] for x in full_papers]
```

```python
pd.Series(ss_targets).value_counts().head(20)
```

```python
unique_ss_targets = list(set(ss_targets))
ct = 0
for target in unique_ss_targets:
    nums = Composition(target).get_el_amt_dict().values()
    if all(x.is_integer() for x in nums):
        ct += 1
print(ct/len(unique_ss_targets))
```

```python
all_els_precs = []
all_elements = [Element.from_Z(i).symbol for i in range(1, 119)]
prec_aliases = {}
index = 0
for el in all_elements:
    new_dict = {}
    new_dict["Element"] = el
    new_dict["Precursors"] = {}
    for rxn in full_papers:
        temp_time = rxn['temp_time_vector']
        if temp_time[index] <= 0:
            continue
        for prec in rxn['precursors']:
            prec_comp = Composition(prec)
            prec_els = list([el for el in prec_comp.get_el_amt_dict().keys() if el not in ['C', 'H', 'O']])
            if el in prec_els:
                if prec_comp.reduced_formula not in new_dict["Precursors"]:
                    new_dict["Precursors"][prec_comp.reduced_formula] = []
                if prec_comp.reduced_formula not in prec_aliases:
                    prec_aliases[prec_comp.reduced_formula] = []
                prec_aliases[prec_comp.reduced_formula].append(prec)
                new_dict["Precursors"][prec_comp.reduced_formula].append(temp_time[index])
    all_els_precs.append(new_dict)
```

```python
all_els_precs_2 = []
prec_aliases_2 = {}
index = 1
for el in all_elements:
    new_dict = {}
    new_dict["Element"] = el
    new_dict["Precursors"] = {}
    for rxn in full_papers:
        temp_time = rxn['temp_time_vector']
        if temp_time[index] <= 0:
            continue
        for prec in rxn['precursors']:
            prec_comp = Composition(prec)
            prec_els = list([el for el in prec_comp.get_el_amt_dict().keys() if el not in ['C', 'H', 'O']])
            if el in prec_els:
                if prec_comp.reduced_formula not in new_dict["Precursors"]:
                    new_dict["Precursors"][prec_comp.reduced_formula] = []
                if prec_comp.reduced_formula not in prec_aliases_2:
                    prec_aliases_2[prec_comp.reduced_formula] = []
                prec_aliases_2[prec_comp.reduced_formula].append(prec)
                new_dict["Precursors"][prec_comp.reduced_formula].append(temp_time[index])
    all_els_precs_2.append(new_dict)
```

```python
for el in all_els_precs:
    if el['Element'] == 'Fe':
        plt.rcParams["figure.figsize"] = [8,5]
        fig, ax = plt.subplots()
        ax.set_xlabel("Solid-State Calcination Temperature (C)")
        labels = []
        for prec in el['Precursors']:
            #if len(el['Precursors'][prec]) > 30:
            if prec in ["Fe2O3", "Fe3O4", "Fe(CO2)2", "Fe(NO3)3", "FePO4"]:
                labels.append("{}, {}".format(max(set(prec_aliases[prec]), key=prec_aliases[prec].count), len(el['Precursors'][prec])))
                sns.kdeplot(el['Precursors'][prec])
        ax.legend(labels, loc="upper left")
```

```python
for el in all_els_precs_2:
    if el['Element'] == 'Fe':
        plt.rcParams["figure.figsize"] = [8,5]
        fig, ax = plt.subplots()
        ax.set_xlabel("Solid-State Sintering Temperature (C)")
        labels = []
        for prec in el['Precursors']:
            #if len(el['Precursors'][prec]) > 10:
            if prec in ["Fe2O3", "Fe3O4", "Fe(CO2)2", "Fe(NO3)3", "FePO4"]:
                labels.append("{}, {}".format(max(set(prec_aliases_2[prec]), key=prec_aliases_2[prec].count), len(el['Precursors'][prec])))
                sns.kdeplot(el['Precursors'][prec])
        ax.legend(labels, loc="upper left")
```

```python
# plotting results
to_search = ["Fe3O4", "Fe2O3", "Fe(NO3)3", "FePO4", "Fe(CO2)2"]
for el in all_els_precs:
    if el['Element'] == 'Fe':
        plt.rcParams["figure.figsize"] = [8,5]
        fig, ax = plt.subplots()
        ax.set_xlabel("Dataset Solid-State Calcination Temperature ($^{\circ}$C)", fontsize=14)
        labels = []
        for i in range(len(to_search)):
            for prec in el['Precursors']:
                #if len(el['Precursors'][prec]) > 30:
                if prec == to_search[i]:
                    labels.append("{}, {}".format(max(set(prec_aliases[prec]), key=prec_aliases[prec].count), len(el['Precursors'][prec])))
                    sns.kdeplot(el['Precursors'][prec])
ax.legend(labels, loc="upper right", fontsize=12)
ax.set_ylabel("Density", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
fig.tight_layout()
fig.savefig('/home/jupyter/CJK/CVAE/ceder_imputed/all_targets/neurips_paper/figures/ss_precursor_calcine_trends.png', dpi=200)
```

```python
# plot sintering temps
to_search = ["Fe3O4", "Fe2O3", "Fe(NO3)3", "FePO4", "Fe(CO2)2"]
for el in all_els_precs_2:
    if el['Element'] == 'Fe':
        plt.rcParams["figure.figsize"] = [8,5]
        fig, ax = plt.subplots()
        ax.set_xlabel("Dataset Solid-State Sintering Temperature ($^{\circ}$C)", fontsize=14)
        labels = []
        for i in range(len(to_search)):
            for prec in el['Precursors']:
                #if len(el['Precursors'][prec]) > 30:
                if prec == to_search[i]:
                    labels.append("{}, {}".format(max(set(prec_aliases[prec]), key=prec_aliases[prec].count), len(el['Precursors'][prec])))
                    sns.kdeplot([x for x in el['Precursors'][prec] if x > 100], cut=3)
ax.legend(labels, loc="upper right", fontsize=12)
ax.set_ylabel("Density", fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
fig.tight_layout()
fig.savefig('/home/jupyter/CJK/CVAE/ceder_imputed/all_targets/neurips_paper/figures/ss_precursor_sinter_trends.png', dpi=200)
```

```python
all_comps_precs_temps = []
for target in set(ss_targets):
    for rxn in full_papers:
        temps = rxn['temp_time_vector'][:4]
        times = rxn['temp_time_vector'][4:]
        if rxn['target'] == target:
            found = False
            for entry in all_comps_precs_temps:
                if entry['Target'] == target and set(entry['Materials']) == set(rxn['precursors']):
                    found = True
                    if temps[0] > 0:
                        entry['Calcine_temps'].append(temps[0])
                    if temps[1] > 0:
                        entry['Sinter_temps'].append(temps[1])
            if not found and (temps[0] > 0 or temps[1] > 0):
                new_entry = {}
                new_entry['Target'] = target
                new_entry['Materials'] = rxn['precursors']
                new_entry['Calcine_temps'] = []
                new_entry['Sinter_temps'] = []
                if temps[0] > 0:
                    new_entry['Calcine_temps'].append(temps[0])
                if temps[1] > 0:
                    new_entry['Sinter_temps'].append(temps[1])
                all_comps_precs_temps.append(new_entry)
```

```python
target = "LiFePO4"
labels = []
plt.rcParams["figure.figsize"] = [8,5]
fig, ax = plt.subplots()
ax.set_xlabel("Solid-State Calcination Temperature (C)")
ax.set_ylabel("Counts")
ax.set_title(target)
bins = np.linspace(200, 1200, 30)
for entry in all_comps_precs_temps:
    if entry['Target'] == target and len(entry['Calcine_temps']) > 2:
        labels.append("{}, {}".format(entry['Materials'], len(entry['Calcine_temps'])))
        sns.kdeplot(entry['Calcine_temps'])
        #plt.hist(entry['Calcine_temps'], bins=bins, alpha=0.3)
ax.legend(labels, loc="upper right")
```
